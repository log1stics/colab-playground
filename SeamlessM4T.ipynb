```python
#@title Download source voice

from yt_dlp import YoutubeDL
import os

yt_url = 'https://www.youtube.com/watch?v=1j0X9QMF--M'
outtmpl = 'yt.m4a'
option = {
        'outtmpl' : outtmpl,
        'format' : 'bestaudio[ext=m4a]/best[ext=m4a]/best',
    }

ydl = YoutubeDL(option)


if os.path.exists(outtmpl):
    os.remove(outtmpl)


result = ydl.download([yt_url])


import torchaudio
resample_rate = 16000
waveform, sample_rate = torchaudio.load(outtmpl)
resampler = torchaudio.transforms.Resample(sample_rate, resample_rate, dtype=waveform.dtype)
resampled_waveform = resampler(waveform)
torchaudio.save('resampled.wav', resampled_waveform, resample_rate)
```

```python
#@title Setup model
import torch
import torchaudio
from seamless_communication.models.inference import Translator

# Initialize a Translator object with a multitask model, vocoder on the GPU.
translator = Translator("seamlessM4T_medium", "vocoder_36langs", torch.device("cuda:0"))
```

```python
#@title Inference
translated_text, wav, sr = translator.predict('resampled.wav', "s2st", 'jpn')

print(translated_text)

torchaudio.save(
    'generated.wav',
    wav[0].cpu(),
    sample_rate=sr,
)

from IPython.display import Audio

Audio('generated.wav')
```
